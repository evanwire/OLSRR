```python
import pandas as pd 
import statsmodels.regression as reg 
import olsrr
import numpy as np 
```

```python
#Here we will generate a dummy dataset for testing the regression features
X = []
Y = []

np.random.seed(12345)

for i in range(50):
    X.append(i * 0.5)
    Y.append(2 + 4 * (i * 0.5) + np.random.uniform(-4, 4, 1)[0])
    
sim_data = pd.DataFrame({'X': X, 'Y': Y})
sim_data.head()
```

```python
#Initialize the model
reg = olsrr.simple_regression('Y ~ X', sim_data)
```
